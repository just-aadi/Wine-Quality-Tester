```python
import pandas as pd
import matplotlib.pyplot as plt

column_name = 'fixed acidity'

plt.figure(figsize=(10, 6))
plt.hist(df[column_name], bins=20, color='skyblue', edgecolor='black')

plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


x_column = 'fixed acidity'
y_column = 'pH'

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df[x_column], df[y_column], color='blue', alpha=0.5)  # Adjust color and transparency as needed

plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f'Scatter Plot: {x_column} vs {y_column}')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='quality', data=df)
plt.title('Distribution of Quality')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='pH', data=df, palette='viridis')
plt.title('Box Plot of pH vs Quality')
plt.xlabel('Quality')
plt.ylabel('pH Value')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
 data_independent= df.drop(columns=['quality'])
 target=df.quality
 from sklearn.model_selection import train_test_split
 X_train,X_test,y_train,y_test= train_test_split(data_independent,target,test_size=0.2)
```

```python
from sklearn import tree
model= tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
model.predict([[7.4,0.7,0,1.9,0.076,11,34,0.9978,3.51,0.56,9.4]])
model.score(X_test,y_test)
```
